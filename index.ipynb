```python
# Import necessary libraries
import re
import requests
import traceback
from random import randint
from time import sleep
from pprint import pprint
from datetime import datetime
import pandas as pd
from bs4 import BeautifulSoup
```

```python
username: str = 'hereugo'
max_race_num: int = 10950

url: str = 'https://data.typeracer.com/pit/'

link: str = url + 'result?id=|tr:{username}|{race_num}'
```

```python
def save_data_to_csv(data: dict, filename: str):
    df = pd.DataFrame(data)
    # open existing csv file and append new data
    try:
        existing_df = pd.read_csv(filename)
        df = pd.concat([existing_df, df], ignore_index=True)
    except FileNotFoundError:
        pass

    df.to_csv(filename, index=False)
```

```python

race_num: int = 174

while True:
    try:
        history = []
        old_race_num: int = race_num

        while race_num != max_race_num:
            try:
                # Get info on race
                response = requests.get(link.format(username=username, race_num=race_num))
                soup = BeautifulSoup(response.text, 'html.parser')

                table: BeautifulSoup = soup.find('table', class_='raceDetails')
                if table is None:
                    raise Exception('No table found')
            
                data = {
                    'race': None,
                    'date': None,
                    'speed': None,
                    'ghost_link': None,
                    'accuracy': None,
                    'rank': None,
                    'rank_total': None
                }

                # row 1: racer name
                # row 2: race number
                # row 3: date
                # row 4: speed (wpm) and link to play again
                # row 5: accuracy
                # row 6: rank
                table_rows: [BeautifulSoup] = table.find_all('tr')

                # skip the first row (racer name)
                
                data['race'] = int(table_rows[1].find_all('td')[1].text)
                data['date'] = datetime.strptime(
                    table_rows[2].find_all('td')[1].text.replace('+0000', '').strip(),
                    '%a, %d %b %Y %H:%M:%S'
                )
                data['speed'] = int(table_rows[3].find_all('td')[1].find('span').text.replace(' WPM', ''))
                data['ghost_link'] = table_rows[3].find_all('td')[1].find('a')['href']
                data['accuracy'] = float(table_rows[4].find_all('td')[1].text.replace('%', '').strip())
                data['rank'] = int(re.findall('\d+', table_rows[5].find_all('td')[1].text)[0])
                data['rank_total'] = int(re.findall('\d+', table_rows[5].find_all('td')[1].text)[1])
            except Exception as e:
                print('Stopped at race: ', race_num)
                print(traceback.format_exc())
                break
            
            history.append(data)
            race_num += 1
        
        save_data_to_csv(history, f'{username}_history.csv')
        print(f'Saving from {old_race_num} to {race_num}')
        sleep(randint(1, 10))
    except Exception as e:
        print('STOPPED:', race_num)
        print(traceback.format_exc())
```
